```python
import pandas as pd
import datetime as dt

import matplotlib.pyplot as plt
import matplotlib
% matplotlib inline
import seaborn as sns
sns.set(style="white")
```

```python
df = pd.read_csv('energy_consumption_data.csv', index_col = 'Timestamp', parse_dates = True)
df.head()
```

```python
df.plot()
```

```python
df['year'] = df.index.year
df['month'] = df.index.month
df['day'] = df.index.day
df['hour'] = df.index.hour
df.head()
```

```python
period = 'year'
facet = sns.FacetGrid(df, hue=period, aspect=4)
facet.map(sns.kdeplot, 'Energy_consumption_value', shade=True)
facet.add_legend()
plt.show()
```

```python
period = 'month'
facet = sns.FacetGrid(df, hue=period, aspect=4)
facet.map(sns.kdeplot, 'Energy_consumption_value', shade=True)
facet.add_legend()
plt.show()
```

```python
period = 'day'
facet = sns.FacetGrid(df, hue=period, aspect=4)
facet.map(sns.kdeplot, 'Energy_consumption_value', shade=True)
facet.add_legend()
plt.show()
```

```python
period = 'hour'
facet = sns.FacetGrid(df, hue=period, aspect=4)
facet.map(sns.kdeplot, 'Energy_consumption_value', shade=True)
facet.add_legend()
plt.show()
```

years and days are similar, months and hours are different
high skew

box plots...

```python
fig = plt.figure(figsize=(14,6))
sns.boxplot(data=df, x='year', y='Energy_consumption_value')
plt.show()
```

```python
fig = plt.figure(figsize=(14,6))
sns.boxplot(data=df, x='month', y='Energy_consumption_value')
plt.show()
```

```python
fig = plt.figure(figsize=(14,6))
sns.boxplot(data=df, x='day', y='Energy_consumption_value')
plt.show()
```

```python
fig = plt.figure(figsize=(14,6))
sns.boxplot(data=df, x='hour', y='Energy_consumption_value')
plt.show()
```

years and days similar

months have clear trend...curve fitting possibility
